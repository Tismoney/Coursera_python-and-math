```python
from math import sqrt
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
```

# Распредление Лапласа

Я выбрал распредление Лапласа. Оно часто встречается в физике и в теории вероятности. Задается она с помощью плотности вероятности
    \begin{equation}
    f(x) = \frac{\alpha}{2}e^-\alpha(x-\beta)
    \end{equation}
Для однозначного определения этого распредления нужны два параметра $\alpha > 0$ - параметр масштаба и $\beta$ - параметр сдвига. Математическое ожидание (среднее значение) такого распредления равно $\beta$, а диспресия (квадрат отклонения) равна $\frac{2}{\alpha^2}$

Нарисуем плотность вероятности и функцию распределения для распредления Лапласа с параметрами $\alpha = 1$ и $\beta = 0$

```python
x = np.arange(-5,5,0.1)
y1 = stats.laplace.pdf(x,loc = 0,scale = 1)
y2 = stats.laplace.cdf(x,loc = 0,scale = 1)
```

## Плотность вероятности

```python
print plt.plot(x, y1, label = 'Probability \ndensity function')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

## Фцункция распредления

```python
plt.plot(x, y2, label = 'Cumulative \ndistribution function')
plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper left')
```

# Проверка Центральной предельной теоремы

## Генерация случаных чисел

```python
def avarage(mass):
    sum = 0
    for i in range(len(mass)):
        sum += mass[i]
    return sum/len(mass)
```

```python
def generate(a, b, n, N):
    sample = np.arange(0, N, dtype = float)
    for i in range(N):
        sample[i] = avarage(stats.laplace.rvs(loc=a, scale=b, size=n, random_state=None))
    return sample
```

## Построение гистограмм

### Анализ проведен на данных размером 5

```python
sample = generate(0, 1, 5, 1000)
plt.hist(sample, normed = True)
x = np.arange(-4,4,0.1)
y3 = stats.norm.pdf(x, 0 , sqrt(2./5))
plt.plot(x, y3)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Отчетливо видно, что два последние гистограммы имею более симметричный вид, чем первые две. Эти эмпирические выводы могут утверждать, что центральная предельная теорема верна.
